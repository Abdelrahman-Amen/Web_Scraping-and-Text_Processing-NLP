```python
import requests
from bs4 import BeautifulSoup
import csv
import os
from google.colab import files
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer,PorterStemmer
import string
```

```python
page=requests.get(f"https://builtin.com/artificial-intelligence")
```

```python
def main(page):

  scr=page.content
  soup=BeautifulSoup(scr,"lxml")
  # matches_details=[]
  Document=soup.find_all("div",{"class":"clearfix text-formatted field field--name-field-p-body field--type-text-long field--label-hidden field__item"})  # to take part of the website
  # print(Document)


  def get_text_info(Document):
      text_info = []
      # Initialize NLTK resources
      nltk.download('punkt')
      nltk.download('wordnet')
      nltk.download('stopwords')
      stop_words = set(stopwords.words('english'))
      lemmatizer = WordNetLemmatizer()
      stemmer = PorterStemmer()
      for item in Document:
          # Find only direct child <p> elements, not nested ones
          paragraphs = item.find_all("p", recursive=False)
          for paragraph in paragraphs:
              text_title = paragraph.text.strip()
              # Tokenization
              tokens = word_tokenize(text_title)
              # Cleaning and Normalization
              tokens = [word.lower() for word in tokens if word.isalnum()]
              # tokens = [word.lower() for word in tokens if word.isalpha()]

              tokens = [word for word in tokens if word not in stop_words]
              # Stemming
              tokens = [stemmer.stem(word) for word in tokens]
              # Lemmatization
              tokens = [lemmatizer.lemmatize(word) for word in tokens]
              text_info.extend(tokens)
      return text_info


  text_data = get_text_info(Document)
  print(text_data)

    # Get unique words
  unique_words = set(text_data)

    # Writing to CSV
  with open('output.csv', 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Text'])
        for text in unique_words:
            if text.strip():  # Check if the text is not empty
                writer.writerow([text])
main(page)
```
